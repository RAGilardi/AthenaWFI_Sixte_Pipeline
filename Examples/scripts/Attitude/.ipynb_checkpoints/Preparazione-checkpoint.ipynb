# 2) Fit da File esempio

```python
file=fits.open('CDFS_lissajous_80ksec.att')
print(file.info())
```

```python
print(file[1].header)
```

```python
print(file[1].data)
```

```python
file_table = table.Table.read(file, memmap=True)

print(file_table)
```

## Estrazione fit attitude

```python
from scipy.optimize import curve_fit

x0 = 53.15
y0 = -27.8

def fit_ra(t, A, dx, a):
    return A * np.cos(dx * t) + x0 + A/a

def fit_dec(t, A, dy):
    return A * np.sin(dy * t) + y0
```

```python
t = file_table[0][:]
ra = file_table[1][:]
dec = file_table[2][:]

#fattore di conversione da sec (8e4) a IU (t)
gain = 8e4 / t[-1]
print(t[-1])
print(len(t))
print('gain=', gain)

step = (t[-1]-t[0])/len(t)
print(step)
```

```python
#fit ra
iguess=[2e-1, 8e-4, 2]
popt_ra, pcov = curve_fit(fit_ra, t, ra, p0=iguess)
print(popt_ra)
#print(popt_ra[2]/x0)

plt.plot(t, ra, 'r.',ms=10)
plt.plot(t, fit_ra(t, *popt_ra),'b')

#plt.xlim([0,5e3])
```

```python
#fit dec

N=1000
#N=len(dec)

iguess=[6e-2, 1.2e-3]
popt_dec, pcov = curve_fit(fit_dec, t[:N], dec[:N], p0=iguess,  bounds=((1e-2,-np.inf), (1e-1,np.inf)))
print(popt_dec)

plt.plot(t, dec, 'r.',ms=10)
plt.plot(t, fit_dec(t, *popt_dec),'b')
#plt.xlim([0,5e4])
```
