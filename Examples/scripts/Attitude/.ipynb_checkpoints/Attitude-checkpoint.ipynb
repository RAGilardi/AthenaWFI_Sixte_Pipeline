# Dati

```python
#RA e DEC
RA=15.670014211242
DEC=-49.293331613352
tstop = 8e4
```

# Codice

```python
import numpy as np
import matplotlib.pyplot as plt

from pylab import rcParams
rcParams['figure.figsize'] = 9, 9

from astropy.io import fits
from astropy import table
```

```python
def func_ra(t, RA):
    A = 3.35250000e-02
    dx = 8.14332248e-04
    a = 0.013525
    #a = 0.046636746347910533*x0
    return A * np.cos(dx * t) + RA + a

def func_dec(t,DEC):
    A = 0.033525
    dy = 0.00114007
    return A * np.sin(dy * t) + DEC
```

```python
#Conversione
gain=0.501137174211506
step=3.069940961538461*gain
t_sec = np.arange(start=0, stop=tstop-step/2, step=step)
t_new = t_sec / gain

ra_new = func_ra(t_new, RA)
dec_new = func_dec(t_new, DEC)

plt.plot(ra_new, dec_new)
```

```python
file=fits.open('blank_80ksec.att')
```

```python
for i in range(len(t_new)):
    file[1].data[i][0]=t_new[i]
    file[1].data[i][1]=ra_new[i]
    file[1].data[i][2]=dec_new[i]
    
file.writeto('attitude_80ksec.att', overwrite=True)
```
